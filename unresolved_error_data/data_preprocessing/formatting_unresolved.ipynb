```python
import pandas as pd
import re
```

```python
unresolved_error = pd.read_csv('./unresolved_error_dataset.csv', index_col=0)
unresolved_error = unresolved_error.rename(columns={'story_truncated':'story_ending'})
unresolved_error
```

```python
unresolved_error['filter'] = unresolved_error.apply(lambda x: x['reveal_sentence'] in x['story'], axis=1)
unresolved_error = unresolved_error.loc[unresolved_error['filter']].reset_index(drop=True)
def remove_blanks(row):
    lines = row['story'].splitlines()
    return ''.join(line for line in lines if line.strip())
unresolved_error['story'] = unresolved_error.apply(remove_blanks, axis=1)
```

```python
def get_prop(row):
    story = row['story']
    truncate = row['story_ending']
    sentences = [x.strip().replace('\n','') for x in story.split('.') if re.search(r'\w', x)!=0]
    truncated_sentences = [x.strip().replace('\n','') for x in truncate.split('.') if re.search(r'\w', x)!=0]
    return len(truncated_sentences)/len(sentences)

def truncate_story(row):
    story = row['story']
    line = row['reveal_sentence']
    print(story.find(line))
    return story[:story.find(line)]
unresolved_error['chunk_prop'] = unresolved_error.apply(get_prop, axis=1)
unresolved_error['story_truncated'] = unresolved_error.apply(truncate_story, axis=1)
```

```python
def write_synthetic_datapoint_to_file(X, y, path, plot_hole_type):
    """
    write a synthetic datapoint to a file.
    :param X: synthetic document
    :param y: synthetic label
    :param path: path to write the file to
    :param plot_hole_type: type of plot hole, will be written at top of document
    :returns: None. file will be written at path. first line will be "plot_hole_type y", and
    rest of the lines will be X.
    """
    with open(path, "w", encoding="utf-8") as synthetic_document_f:
        synthetic_document_f.write(f"{plot_hole_type} {y}\n")
        synthetic_document_f.write(X)
```

```python
for i,row in unresolved_error.iterrows():
    path = f"./test/test_unresolved{i}.txt"
    X = row['story_truncated']
    y = row['chunk_prop']
    write_synthetic_datapoint_to_file(X,y,path=path,plot_hole_type="unresolved")
```
